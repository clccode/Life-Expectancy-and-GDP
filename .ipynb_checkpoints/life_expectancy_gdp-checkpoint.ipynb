### Import the relevant libraries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

### Import the data

```python
df = pd.read_csv('all_data.csv')
# inspect the first few rows
df.head()
```

```python
# Which countries are included?
df['Country'].unique()
```

```python
# Each country as a dataframe
cl = df[df['Country'] == 'Chile']
cn = df[df['Country'] == 'China']
de = df[df['Country'] == 'Germany']
mx = df[df['Country'] == 'Mexico']
us = df[df['Country'] == 'United States of America']
zw = df[df['Country'] == 'Zimbabwe']
```

```python
# Life expectancy over time in Chile
plt.plot(cl['Year'], cl['Life expectancy at birth (years)'], color='blue', marker='o')
plt.title('Life Expectancy in Chile 2000-2015')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.grid(True)
plt.show()
```

```python
# Zimbabwe?
plt.plot(zw['Year'], zw['Life expectancy at birth (years)'], color='green', marker='o')
plt.title('Life Expectancy in Zimbabwe 2000-2015')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.grid(True)
plt.show()
```

```python
# US?
plt.plot(us['Year'], us['Life expectancy at birth (years)'], color='red', marker='o')
plt.title('Life Expectancy in the US 2000-2015')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.grid(True)
plt.show()
```

```python
#China?
plt.plot(cn['Year'], cn['Life expectancy at birth (years)'], color='purple', marker='o')
plt.title('Life Expectancy in China 2000-2015')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.grid(True)
plt.show()
```

```python
#Germany?
plt.plot(de['Year'], de['Life expectancy at birth (years)'], color='orange', marker='o')
plt.title('Life Expectancy in Germany 2000-2015')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.grid(True)
plt.show()
```

```python
#GDP in Chile?
plt.plot(cl['Year'], cl['GDP'], color='red', marker='o')
plt.title('Chile: GDP 2000-2015')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.grid(True)
plt.show()
```
